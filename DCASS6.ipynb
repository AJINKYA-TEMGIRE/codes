```python
from PIL import Image
import numpy as np
import tensorflow as tf
import tensorflow_hub  as hub

def load(path):
    img = Image.open(path).convert("RGB")
    img = img.resize((256 , 256))
    img = np.array(img) / 255.0
    img = img.astype(np.float32)
    return tf.image.convert_image_dtype(img , tf.float32)[tf.newaxis , ...]


content_image = load("cat.png")
style = load("styled.png")

model = hub.load("https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2")

styled_image = model(tf.constant(content_image) , tf.constant(style))[0]

import matplotlib.pyplot as plt

plt.imshow(tf.squeeze(styled_image))
```
